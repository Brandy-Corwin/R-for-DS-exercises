```r
library(ggplot2)
library(tidyverse)
user2025 <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-04-29/user2025.csv')
```

```r
options(repr.plot.width = 20, repr.plot.height = 8)
user2025 |>
ggplot(aes(x=room, color=session, fill=session)) +
geom_bar() +
labs(
        title = "Types of Sessions in Each Room",
        y = "Count"
    ) + 
    theme(plot.title = element_text(size = 30),
    axis.text = element_text(size = 13),
    axis.title = element_text(size = 15),
    legend.title = element_text(size=15),
    legend.text = element_text(size=13)
    )
```
