```r
library(ggplot2)
library(tidyverse)
library(grid)
nsf_terminations <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-05-06/nsf_terminations.csv')
```

```r
options(repr.plot.width = 25, repr.plot.height = 10)
nsf_terminations |>
    ggplot(aes(x = org_state, y = usaspending_obligated, color = org_state)) +
    geom_boxplot() +
    facet_wrap(~ in_cruz_list) +
    labs(
        title = "Amount of Money that was Terminated From Grants In Each State Based on If the Grant was in Cruz's List",
        x = "State",
        y = "Amount of Money"
    ) + 
    theme(plot.title = element_text(size = 30),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 15),
    legend.title = element_text(size=15),
    legend.text = element_text(size=13)
    ) +
    theme(panel.spacing = unit(2, "lines"))
```
