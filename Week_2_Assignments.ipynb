a. Write down what you think the output will look like, then check if you were correct, and describe what group_by() does.
```r
df |>
  group_by(y)
```

The output will be the table. It will state that it has been grouped but not look any different. group_by() doesn't change the data, but future operartions will be done by whatever you grouped by.

````r
df |>
  group_by(y)
````

b. Write down what you think the output will look like, then check if you were correct, and describe what arrange() does. Also, comment on how it’s different from the group_by() in part (a).
```r
df |>
  arrange(y)
```

arrange(y) will reorder the rows so that all values in y are ordered alphabetically in this case. arrange() reorders the table from smallest to largest or alphabetically depending on the type of column chosen. As opposed to group_by, this does chage what the output table looks like.

````r
df |>
  arrange(y)
````

c. Write down what you think the output will look like, then check if you were correct, and describe what the pipeline does.
```r
df |>
  group_by(y) |>
  summarize(mean_x = mean(x))
```

The output will have one column of unique y values with another being what the mean of the x columns were for each value of y. The pipline is saying to take the output of that line and use it as input for the next line, so it takes the dataframe, groups by y, then summarises the mean of x using the group by.

````r
df |>
  group_by(y) |>
  summarize(mean_x = mean(x))
````

d. Write down what you think the output will look like, then check if you were correct, and describe what the pipeline does. Then, comment on what the message says.
```r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))
```

It does the same as in the previous part, but groups by y and z instead of just y. The pipeline takes the dataframe, groups by y and z, then summarises the mean of x using the group by and removes the y grouping. The message is saying that the summarise removes one grouping level.

````r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))
````

e. Write down what you think the output will look like, then check if you were correct, and describe what the pipeline does. How is the output different from the one in part (d)?

```r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x), .groups = "drop")
```

The output look the same as in the previous part, but the message is gone because we dropped all groupings so the table is no longer grouped. The pipeline takes the dataframe, groups by y and z, then summarises the mean of x using the group by and removes both groupings.

````r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x), .groups = "drop")
````

f. Write down what you think the outputs will look like, then check if you were correct, and describe what each pipeline does. How are the outputs of the two pipelines different?
```r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))

df |>
  group_by(y, z) |>
  mutate(mean_x = mean(x))
```

The first section is exactly part d. the second will look like the original table but with an extra column that has the means of x for each y,z grouping. The first pipeline takes the dataframe, groups by y and z, then summarises the mean of x using the group by and removes the y grouping. The second pipeline takes the dataframe, groups by y and z, then adds a column for the mean of x using the mutate function keeping the table grouped by both y and z.

````r
df |>
  group_by(y, z) |>
  summarize(mean_x = mean(x))

df |>
  group_by(y, z) |>
  mutate(mean_x = mean(x))
````
